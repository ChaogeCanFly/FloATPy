```python
def merge_dicts(old1,old2):
    new = {}
    for i in range(3):
        key = old1.keys()[i]
        v1 = old1.values()[i][0]
        v2 = old2.values()[i][0]
        e1 = old1.values()[i][1]
        e2 = old2.values()[i][1]
        err = (e1**2+e2**2)**0.5 + np.std([v1,v2])
        new[key] = [(v1+v2)/2., err]
    return new

# rr = merge_dicts(Mt_rr7,Mt_rr_7)
def plotter(dat,fmt,label,fillstyle='full'):
    plt.errorbar(dat.keys(),[x[0] for x in dat.values()],
        yerr=[x[1] for x in dat.values()],fmt=fmt,capsize=3,label=label,fillstyle=fillstyle)

fig = plt.figure(figsize=(3,3),dpi=200)
plotter(Mt['s1'],'ko','$M_t$')
plotter(Mtv['s1'],'ko','$M_{t,v}$',fillstyle='none')
plotter(Mg['s1'],'k^','$M_g$')
plotter(Mtau['s1'],'ks',r'$M_\tau$')

plt.xlabel(r'$M_c$')
# plt.ylabel(r'$M_{t,v}\approx\sqrt{|R_{22}|}/\bar{c}$')
plt.legend(fontsize=10)
plt.xlim([0,2.1])
plt.tight_layout()
```

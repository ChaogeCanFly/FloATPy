```python
# Integrated profiles
# 1D integration in y using midpoint rule:

def get_Mtau(directory,flist):
    
    mat = np.zeros([len(flist),2])
    for i,fname in enumerate(flist):
        tID = fname.split('/')[-1]
        tID = tID.split('_')[-1]
        tID = int(tID.split('.')[0])
        time = nml.read_time(directory,tID)

        # Get dtheta
        idx = np.argmin(abs(tlist-time))
        dtheta = dlist[idx]
        eta = y/dtheta
        
        # Get centerline
        ic,yc = get_centerline(directory,-y,tID)

        # Get cbar:
        cbar = np.fromfile( directory + 'shearlayer_cbar_%04d'%(tID)+'.dat',dtype=float, count=-1, sep=' ')
        
        # Get shear stress
        R11,R12,R13,R22,R23,R33 = get_Rij(fname)
        Mtau = np.sqrt(abs(R12))/cbar
        plt.plot(eta,Mtau)
        
        # integrate
        mat[i,0] = time
        mat[i,1] = np.amax(Mtau)
        
    # sort by time
    isort = np.argsort(mat[:,0]) #sort time
    return mat[isort,:]


fig = plt.figure(dpi=100)
dat = get_Mtau(directory,flist[3:])
plt.xlim([-20,20])
# plt.plot(dat[:,0]/inp.du,dat[:,1],'*')
```
